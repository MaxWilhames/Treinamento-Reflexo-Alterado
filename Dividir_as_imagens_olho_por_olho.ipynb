```python
import cv2
import os
from google.colab import drive

# Monta o Google Drive
drive.mount('/content/drive')

# Pasta onde estão suas imagens
pasta = "/content/drive/MyDrive/datset_roboflow"

# Pasta de saída (se não existir, cria)
saida = "/content/drive/MyDrive/datset_roboflow_divididas"
os.makedirs(saida, exist_ok=True)

# Lista todos os arquivos da pasta
for arquivo in os.listdir(pasta):
    caminho = os.path.join(pasta, arquivo)

    # Verifica se é arquivo de imagem
    if arquivo.lower().endswith(('.jpg', '.jpeg', '.png')):
        imagem = cv2.imread(caminho)

        if imagem is None:
            print(f"Erro ao carregar {arquivo}")
            continue

        # Obtém dimensões
        altura, largura, _ = imagem.shape
        meio = largura // 2

        # Divide
        esquerda = imagem[:, :meio]
        direita = imagem[:, meio:]

        # Salva
        nome_base = os.path.splitext(arquivo)[0]
        cv2.imwrite(os.path.join(saida, f"{nome_base}_esquerda.jpg"), esquerda)
        cv2.imwrite(os.path.join(saida, f"{nome_base}_direita.jpg"), direita)

        print(f"{arquivo} processado com sucesso!")

print("✅ Todas as imagens foram processadas e salvas em:", saida)
```
